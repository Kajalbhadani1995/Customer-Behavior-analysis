## Data Cleaning & Transforming

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns= {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
df.info()
```

```python
# Step 1: Calculate average rating per product category
category_avg_rating = (
    df.groupby("item_purchased")["review_rating"]
    .mean()
)

# Step 2: Fill null review ratings using category average
df["review_rating"] = df["review_rating"].fillna(
    df['item_purchased'].map(category_avg_rating)
)

# Step 3: Fill any remaining nulls with overall average rating
overall_avg_rating = df["review_rating"].mean()

df["review_rating"] = df["review_rating"].fillna(overall_avg_rating)

# Check
print(df["review_rating"].isnull().sum())
```

```python
# age_group column
labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"] = pd.qcut(df["age"], q=4, labels=labels)
```

```python
df[["age", "age_group"]].head(5)
```

```python
# Create a mapping dictionary for purchase frequency to days
frequency_mapping = {
    "Daily": 1,
    "Weekly": 7,
    "Fortnightly": 14,
    "Monthly": 30,
    "Quarterly": 90,
    "Bi-Weekly": 14,
    "Annually": 365,
    "Every 3 Months" : 90
}

# Create a new column using the mapping
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```
